```python
#Ex. 1
#From https://medium.com/quick-code/absolute-beginners-guide-to-slaying-apis-using-python-7b380dc82236
import requests
#request = requests.get('http://api.open-notify.org')
#print(request.text)

people = requests.get('http://api.open-notify.org/astros.json')
print(people.text)
```

```python
people_json  = people.json()
#Similar to people.text
#print(people_json)

#To print the  number of people in space
print("Number of people in space:",people_json['number'])

#To print the names of people in space using a for loop
for p in people_json['people']:
    print(p['name'], 'is in: ', p['craft'])
```

```python
#Ex.2
#Adapted from: http://open-notify.org/Open-Notify-API/ISS-Location-Now/
import requests
import json

response = requests.get("http://api.open-notify.org/iss-now.json")
print(response.text)
pos = response.json()
print('Timestamp is: ', pos['timestamp'])
print('Latitude is: ', pos['iss_position']['latitude'])
print('Longitude is: ', pos['iss_position']['longitude'])
```

```python
import datetime
print(datetime.datetime.fromtimestamp(int(pos['timestamp'])))
```

```python
#Ex.3
#Note: Latitude and longitude can be directly added to the end point as shown below
#response = requests.get("http://api.open-notify.org/iss-pass.json?lat=40.71&lon=-74")
#In this example, we will work with parameters and use the latitude and longitude of New York
base_url = 'http://api.open-notify.org/'
newyork = {'lat': 40.71, 'lon': -74 }
response = requests.get(base_url + 'iss-pass.json', params=newyork)
r= response.json()
print(r)
```

```python
import datetime
for p in r['response']:
    #convert from UNIX time
    date = datetime.datetime.fromtimestamp(int(p['risetime']))
    print(date)
```
