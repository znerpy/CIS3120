```python
#Source: Adapted from FreeCodeCamp.org

import requests
from bs4 import BeautifulSoup

r = requests.get("https://www.whitehouse.gov//briefing-room/")
src = r.content
soup = BeautifulSoup(src, 'html.parser')

#store urls in a Python list called urls and initialized with []
urls = []

#Find all h2 tags in the page and within each one find the a tags
for h2_tag in soup.find_all("h2"):
    a_tag=h2_tag.find('a')
    #some tags are None and will give an error in the .attrs statement
    #print(a_tag)
    if a_tag == None:
        pass
    else:
        urls.append(a_tag.attrs['href'])
print(urls)
print("There are ", len(urls), "links on the page")
```

```python
#Source: Adapted from FreeCodeCamp.org

import requests
from bs4 import BeautifulSoup

r = requests.get("https://www.whitehouse.gov//briefing-room/")
src = r.content
soup = BeautifulSoup(src, 'html.parser')

#store urls in a Python list called urls and initialized with []
headline_list = []

#Find all h2 tags in the page and within each one find the a tags
for h2_tag in soup.find_all("h2"):
    a_tag=h2_tag.find('a')
    #some tags are None and will give an error in the .attrs statement
    #print(a_tag)
    if a_tag == None:
        pass
    else:
        headline_list.append(a_tag.text)
#print(headline_list) prints all at once
#To print one at the time
print("Here are the headlines: \n ")
for e in headline_list:
    print(e)
print("There are ", len(headline_list), "headlines on the page")
```

```python
# Web scraping IMDb table
import requests
from bs4 import BeautifulSoup
 
url = 'http://www.imdb.com/chart/top'
r = requests.get(url)
soup = BeautifulSoup(r.text)

print('This is the title:', soup.title.text)
mo_table = soup.find('table')
#print(mo_table)

#get headers and put them in a list
headers = []
#Go through first row and get headers
for header in mo_table.find("tr").find_all('th'):
    headers.append(header.text)
print('These are the headers: \n', headers)
```

```python
#Movies is a ist of dictionaries that keeps table info
movies = []

#There are three nested loops below
#First for loop goes through rows except first one
for row in mo_table.find_all('tr')[1:]:
    values = [] #list to hold row values in each row
    #Second for loop goes through columns
    for col in row.find_all('td'):
        values.append(col.text)
    if values: #Not empty - skips empty values
        mo_dict = {} #Empty dictionary to store row values
        #Third for loop connects values to labels from headers
        for i in range(len(values)):
            mo_dict[headers[i]]=values[i]
    movies.append(mo_dict) #adds dict to movies list
print(movies)
print("The number of movies is: ", len(movies))
```

```python
#Now, we will calculate the average ratings for all the movies
trt = 0  #Total ratings trt is initialized to zero
#Go through each row and extract the rating
for movie in movies:
    #rt is the rating for the current movie
    #replace removes formatting characters
    rt=movie['IMDb Rating'].replace('\n','')
    trt+= float(rt) #accumulates rating after converting to float
    
#To calculate the average, divide total by number of movies
rt_avg= trt/len(movies)
print('The average rating of all the movies is: ', rt_avg)
```

```python
#To format the average to two decimals, use f in print
format_rtavg = "{:.2f}".format(rt_avg)
print('The average rating of all the movies with only two decimals is: ', format_rtavg)
```
