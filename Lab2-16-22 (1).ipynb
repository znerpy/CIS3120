```python
import requests
from bs4 import BeautifulSoup

r = requests.get("http://www.google.com")
src=r.content
soup = BeautifulSoup(src,'html.parser')
soup
```

```python
print('This is the soup object: \n', soup)
```

```python
links=soup.find_all("a")
#Print links to see that looks like a list
print(links)
```

```python
print(type(links))
```

```python
#Calculate the number of links with len
count=len(links)
print('There are a total of ', count, "links in this page")
```

```python
#Print each link separately
for link in links:
    print(link)
```

```python
#Print each link separately
for link in links:
    print(link.attrs['href'])
```

```python
#Find the link with About in the text and print its url
for link in links:
    if "About" in link.text:
        print(link.attrs['href'])
```

```python
#Ex.2: WhiteHouse Briefing Room Scraping 
#Source: Adapted from FreeCodeCamp.org

import requests
from bs4 import BeautifulSoup

r = requests.get("https://www.whitehouse.gov//briefing-room/")
src=r.content
soup = BeautifulSoup(src,'html.parser')

#store urls in a Python list called urls and initialized with []
urls = []

lh2tags = soup.find_all("h2")

#Find all h2 tags in the page and within each one find the a tags
# e is an element in the lh2tag list and allows to iterate through the list, one element at the time
# each e element is an h2 tag in the page, with a tags with links (called h2_tag in the screenshots in the slides)
for e in lh2tags:
    a_tag=e.find('a')
    #some tags are None and will give an error in the .attrs statement
    #print(a_tag)
    if a_tag ==None:
        pass
    else: 
        print(a_tag.attrs['href'])
        urls.append(a_tag.attrs['href'])
```

```python
print(urls)
```
