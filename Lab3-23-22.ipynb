```python
#Ex1: Pandas DataFrames
#Source: https://www.dataquest.io/blog/pandas-python-tutorial/

import pandas as pd
reviews = pd.read_csv("Data/ign.csv")
```

```python
reviews.head()
```

```python
print(type(reviews))
```

```python
#removing the first column (all rows :, column 1 onwards 1:)
reviews = reviews.iloc[:,1:]
reviews.head()
```

```python
print('This is the result of reviews.shape: ', reviews.shape)
```

```python
reviews.info()
```

```python
reviews['score'].describe()
```

```python
reviews['score_phrase'].describe()
```

```python
#Sort by platform in descending order
reviews.sort_values("platform", ascending = False)
```

```python
reviews.head(10)
```

```python
#Filter by Editors choice 
eds_choice = reviews["editors_choice"] == 'Y'
reviews[eds_choice].head()
```

```python
reviews.head()
```

```python
#Show titles chosen by editors with scores less than 9
lo_score = reviews["score"] < 9.0
reviews[eds_choice & lo_score].head()
```

```python
#Count the titles launched each year
reviews["release_year"].value_counts()
```

```python
#Compute max and min of scores
print('max is:',reviews["score"].max())
print('min is:', reviews["score"].min())
```

```python
#Compute max, min, average and standard deviation of scores
print('Maximum score is: ', reviews.score.max())
print('Minimum score is: ', reviews.score.min())
print('Average score is: ', reviews.score.mean())
print('Stdev of score is: ', reviews.score.std())
```
